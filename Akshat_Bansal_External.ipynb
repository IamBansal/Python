```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv("/content/iris.data.csv")
```

```python
df=df.replace({'Iris-setosa':{'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2}})
df
```

```python
df.isnull()
```

```python
df.rename(columns= {'5.1': 'sepal.length', '3.5' : 'sepal.width', '1.4' : 'petal.length', '0.2' : 'petal.width', 'Iris-setosa' : 'variety'}, inplace = True)
```

```python
df
```

```python
features=['sepal.length','sepal.width','petal.length','petal.width']
y=df.variety
x=df[features]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,random_state=42,test_size=0.3)
```

```python
rf=RandomForestClassifier(n_estimators=10,criterion="entropy")
```

```python
rf.fit(X_train,Y_train)
```

```python
y_predict = rf.predict(X_test)
```

```python
accuracy=accuracy_score(y_predict,Y_test)
accuracy
```

```python
X_new=np.array([[5.9,3.0,5.1,1.8]]).reshape(1,-1)
```

```python
pred=rf.predict(X_new)
```
